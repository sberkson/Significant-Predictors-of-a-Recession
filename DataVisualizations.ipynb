```python
# Import libraries

import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in data
data = pd.read_csv(os.path.join(os.getcwd() + '/Data/Master_Data.csv'))

# Drop index and set DATE to datetime dtype
data = data.drop(columns=['Unnamed: 0'])
data["DATE"] = pd.to_datetime(data["DATE"])

# Check data to ensure its good to go
print("Data shape: ", data.shape)
print("Data columns: ", data.columns)
print("Data info: \n", data.info())
```

# Mortgage Debt to Income Ratio 

```python

fig, ax = plt.subplots()

# Initialize the bottom position for the bars
bottom_position = 0


# Plot the bars for 'Recession' equal to 0 in blue
ax.bar(data[data['Recession'] == 0]['DATE'].dt.year, data[data['Recession'] == 0]['MortgageDebtIncomePercent'], color='#a2a471', label='Not in Recession')

# Plot the bars for 'Recession' equal to 1 in red, starting from the bottom
ax.bar(data[data['Recession'] == 1]['DATE'].dt.year, data[data['Recession'] == 1]['MortgageDebtIncomePercent'], color='#2e5543', label='In Recession', bottom=bottom_position)

# Customize labels and titles
ax.set_xlabel('Year', color = '#ffa53b')
ax.set_ylabel('Mortgage Debt to Income Ratio', color = '#ffa53b')
ax.set_title('Average Monthly Mortgage Debt to Income Ratio by Year', color = '#ffa53b')

# Set background to #8e714e
ax.set_facecolor('#8e714e')
# Set entire plot background to #8e714e
fig.set_facecolor('#8e714e')
# Set legend background to #8e714e
ax.legend(facecolor='#8b8e4e')


plt.savefig(os.path.join(os.getcwd() + '/Visualizations/MortgageDebtIncomePercent.png'))
# Show the plot
plt.show()
```

# Poverty Rate by Year

```python
fig, ax = plt.subplots()

# Plot All_Pct, Male_Pct, Female_Pct by year
ax.plot(data['DATE'].dt.year, data['Male_Pct'], color='#a2a471', label='Male-only led household')
ax.plot(data['DATE'].dt.year, data['Female_Pct'], color='#2e5543', label='Female-only led household')
# Set background to #8e714e
ax.set_facecolor('#8e714e')
# Set entire plot background to #8e714e
fig.set_facecolor('#8e714e')
# Set legend background to #8e714e
ax.legend(facecolor='#8b8e4e')

plt.xlabel('Year', color = '#ffa53b')
plt.ylabel('Percent of Households in Poverty', color = '#ffa53b')
plt.title('Percent of Households in Poverty by Year and Family Type', color = '#ffa53b')
plt.savefig(os.path.join(os.getcwd() + '/Visualizations/PovertyByFamilyType.png'))

plt.show()
```
